```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("/content/user_activity_insights/order_patterns_and_preferences/customization_add_ons/part-00000-091dbaae-fa37-4951-84ec-3fe7b933de68-c000.csv")
print(df.head())
print("\nMissing Values:\n", df.isnull().sum())
df.columns = df.columns.str.lower()
df["category"] = df["category"].astype("category")
df["itemname"] = df["itemname"].astype("str")
fig, axes = plt.subplots(3, 2, figsize=(15, 18))
sns.barplot(x="category", y="total_quantity", data=df, ax=axes[0, 0])
axes[0, 0].set_title("Total Quantity per Category")
axes[0, 0].set_xlabel("Category")
axes[0, 0].set_ylabel("Total Quantity")
axes[0, 0].tick_params(axis='x', rotation=45)
sns.barplot(x="itemname", y="total_orders", data=df, ax=axes[0, 1])
axes[0, 1].set_title("Total Orders per Itemname")
axes[0, 1].set_xlabel("Item Name")
axes[0, 1].set_ylabel("Total Orders")
axes[0, 1].tick_params(axis='x', rotation=90)
category_quantity = df.groupby("category")["total_quantity"].sum()
category_quantity.plot(kind="pie", autopct='%1.1f%%', ax=axes[1, 0])
axes[1, 0].set_title("Total Quantity Distribution by Category")
axes[1, 0].set_ylabel('')
sns.scatterplot(x="total_quantity", y="total_orders", data=df, ax=axes[1, 1])
axes[1, 1].set_title("Total Quantity vs Total Orders")
axes[1, 1].set_xlabel("Total Quantity")
axes[1, 1].set_ylabel("Total Orders")
correlation_matrix = df[["total_quantity", "total_orders"]].corr()
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", ax=axes[2, 0])
axes[2, 0].set_title("Correlation Between Total Quantity and Total Orders")
sns.boxplot(x="category", y="total_quantity", data=df, ax=axes[2, 1])
axes[2, 1].set_title("Distribution of Total Quantity by Category")
axes[2, 1].tick_params(axis='x', rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("/content/user_activity_insights/user_demographics_and_segmentation/behavioural_segmentation/part-00000-d9f1b58d-f6d1-4112-8d97-39bfefac5c48-c000.csv")
print("Missing Values:\n", df.isnull().sum())
print("\nDataset Overview:\n", df.head())
sns.set_style("whitegrid")
plt.figure(figsize=(12,6))
sns.barplot(x="segment", y="total_purchases",data=df, palette="viridis", legend=False)
plt.title("Total Purchases Per Segment")
plt.xlabel("Segment")
plt.ylabel("Total Purchases")
plt.xticks(rotation=0)
plt.show()
plt.figure(figsize=(12,6))
sns.boxplot(x="segment", y="total_spent", data=df, palette="viridis", legend=False)
plt.title("Total Spent Per Segment")
plt.xlabel("Segment")
plt.ylabel("Total Spent")
plt.xticks(rotation=0)
plt.show()
plt.figure(figsize=(12,6))
sns.scatterplot(x="sessions_count", y="total_spent", hue="segment", data=df, palette="viridis", alpha=0.7)
plt.title("Sessions Count vs. Total Spent")
plt.xlabel("Sessions count")
plt.ylabel("Total Spent")
plt.legend(title="Segment")
plt.show()
plt.figure(figsize=(8,8))
df["segment"].value_counts().plot.pie(
    autopct=lambda p: f'{p:.1f}%',
    colors=sns.color_palette("pastel"),
    startangle=90,
    wedgeprops={'edgecolor': 'black'}
)
plt.title("User Segment Distribution")
plt.ylabel("")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
file_path = "/content/user_activity_insights/user_engagement_and_behavior/session_duration_frequency/part-00000-bf244305-1135-47cd-9b75-2c305d1f0e24-c000.csv"
df = pd.read_csv(file_path)
df.columns = df.columns.str.lower()
df["session_start"] = pd.to_datetime(df["session_start"], errors = 'coerce')
df["session_end"] = pd.to_datetime(df["session_end"], errors = 'coerce')
df.dropna(subset=["session_start","session_end" ], inplace=True)
df = df.dropna(subset=["session_duration"])
plt.figure(figsize=(12,6))
sns.histplot(df["session_duration"], bins=30, kde=True, color="skyblue")
plt.title("Session Duration Distribution", fontsize=14)
plt.xlabel("Session Duration (minutes)", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.show()
plt.figure(figsize=(12,6))
sns.boxplot(x=df["session_duration"], color="lightcoral")
plt.title("Session Duration Boxplot", fontsize=14)
plt.ylabel("Session Duration (minutes)", fontsize=12)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
file_path = "/content/user_activity_insights/user_engagement_and_behavior/session_duration_frequency/part-00000-bf244305-1135-47cd-9b75-2c305d1f0e24-c000.csv"
df = pd.read_csv(file_path)
df.columns = df.columns.str.lower()
df["session_start"] = pd.to_datetime(df["session_start"], errors = 'coerce')
df["session_end"] = pd.to_datetime(df["session_end"], errors = 'coerce')
df.dropna(subset=["session_start","session_end" ], inplace=True)
df = df.dropna(subset=["session_duration"])
df["session_duration"] = (df["session_end"] - df["session_start"]).dt.total_seconds() / 60
plt.figure(figsize=(12,6))
sns.histplot(df["session_duration"], bins=30, kde=True, color="skyblue")
plt.title("Session Duration Distribution", fontsize=14)
plt.xlabel("Session Duration (seconds)", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.show()
plt.figure(figsize=(12,6))
sns.boxplot(x=df["session_duration"], color="lightcoral")
plt.title("Session Duration Boxplot", fontsize=14)
plt.ylabel("Session Duration (seconds)", fontsize=12)
plt.show()
```
